```python
A= input("Enter the first list of numbers separated by commas: ")
B= input("Enter the second list of numbers separated by commas: ")
A = [int(x) for x in A.split(",")]
B = [int(x) for x in B.split(",")]
```

```python
print(A)
print(B)
```

```python
def check_in_distinct(A, B):
    C = []
    for i in B :
        if A[i] not in C:
            C.append(A[i])
    return C
print(check_in_distinct(A, B))

```

```python
def check(A, B):

    while A and B:
        if A[0] == B[0]:
            A.pop(0)
            B.pop(0)
            return check(A, B)
        else:
            return False
    return False
```
