```python
A= input("Enter the first list of numbers separated by commas: ")
B= input("Enter the second list of numbers separated by commas: ")
A = [int(x) for x in A.split(",")]
B = [int(x) for x in B.split(",")]
```

```python
print(A)
print(B)
```

```python
def distinct(A, B):
    global C
    C = []
    for i in A:
        if i not in B:
            C.append(i)
    for j in B:
        if j not in A:
            C.append(j)
    return C

print(distinct(A, B))
```
